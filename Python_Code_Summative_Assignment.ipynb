```python
patient_BMIs = []
current_patient = []
BMI_sum_list = []

def Calc_BMI(pl_height, pl_weight):
    return ((pl_weight/pl_height/pl_height) * 703)

while True:
    patient_ID = input("Enter the patient's ID number: ")
    try:
        if int(patient_ID) < 0:
            break
    except ValueError:
        print("Please enter a number.")
        continue
    patient_name = input("Enter the patient's name: ")
    
    
    good_input = False
    while not good_input:
        try:
            patient_height = input("Please enter the patient's height in inches: ")
            if float(patient_height) > 0:
                good_input = True
            else:
                print("Please enter a positive number for height in inches.")
        except ValueError:
            print("Please enter a valid value.")
            continue
            
    good_input = False
    while not good_input:
        try:
            patient_weight = input("Please enter the patient's weight in pounds: ")
            if float(patient_weight) > 0:
                good_input = True
            else:
                print("Please enter a positive number for weight in pounds.")
        except ValueError:
            print("Please enter a valid number.")
            continue
            
    p_ID = int(patient_ID)
    p_height = float(patient_height)
    p_weight = float(patient_weight)
    
    current_patient = [p_ID, patient_name, p_height, p_weight] 
    patient_BMIs.append(current_patient)

    
i = 0
print("\n")
print("\t\t\t >------ Patient Report <------")
for patient in patient_BMIs:
    current_BMI = Calc_BMI(patient_BMIs[i][2], patient_BMIs[i][3])
    print(" ")
    print("Patient Number: \t" + str(patient_BMIs[i][0]))
    print("Patient Name:   \t" + str(patient_BMIs[i][1]))
    print("Patient Height: \t" + str(patient_BMIs[i][2]) + " inches")
    print("Patient Weight: \t" + str(patient_BMIs[i][3]) + " lbs")
    print("Patient BMI:    \t" + "{:.1f}".format(current_BMI))
    if current_BMI < 18.5:
        print("BMI Category:   \tUnderweight")
    elif current_BMI >= 18.5 and current_BMI < 25.0:
        print("BMI Category:   \tHealthy weight")
    elif current_BMI >= 25.0 and current_BMI < 30.0:
        print("BMI Category:   \tOverweight")
    else:
        print("BMI Category:   \tObese")
    print(" ")
    BMI_sum_list.append(current_BMI)
    i+=1
Laura 

BMI_sum = sum(BMI_sum_list)
BMI_list_length = len(patient_BMIs)
patient_BMI_average = BMI_sum / BMI_list_length
BMI_max = max(BMI_sum_list)
BMI_min = min(BMI_sum_list)
    
print("\n")
print("Maximum BMI in Group:\t" + "{:.1f}".format(BMI_max))
print("Minimum BMI in Group:\t" + "{:.1f}".format(BMI_min))
print("Average BMI in Group:\t" + "{:.1f}".format(patient_BMI_average))


```
